## Reducing image information with PCA
We use the PCA class of the `scklearn.decomposition` module to display reconstructed versions of an image with fewer principal components than the total. To do this, follow these steps:

```python
from sklearn.decomposition import PCA
from pylab import *
from skimage import data, io, color
from skimage.color import rgb2gray

coke = plt.imread('coca_cola.jpg')
coke_gray = rgb2gray(coke)

plt.imshow(coke_gray,cmap='gray')
xlabel('Original Image')
print(coke_gray.shape)

for i in range(1,5):
    plt.figure()
    n = 4**i
    # Definimos el número de componentes a emplear para cada iteracion
    pca = PCA(n_components = n)
    
    # Aplicamos el modelo PCA a la matriz X extrayendo las componentes en el número decidido n
    coke_gray_pca = pca.fit_transform(coke_gray)
    
    #Reconstruimos la iamgen a partir de las componentes extraidas con PCA.
    coke_gray_restored = pca.inverse_transform(coke_gray_pca)
    plt.imshow(coke_gray_restored,cmap='gray')
    plt.title('Nº componentes: %d' %n)
```
