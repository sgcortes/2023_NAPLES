## Matplotlib examples
Matplotlib is a graphic package for Python. 
[Matplotlib](https://matplotlib.org/)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
x = np.linspace(0,2*np.pi,20)
print(x)
y = np.sin(x)
print(y)
plt.figure()
plt.plot(x,y);
plt.grid('on')
```

```python
plt.plot(x,y)
plt.axis([1,5,-0.5,1]) # Limits x-axis and "y" axis
```

## Pandas
Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool buit on top of Python.
[Pandas](https://pandas.pydata.org/)

### Dataframe:
Pandas is a library for data manipulation whihc allows to access data in different formats and offers indexing data addition and deleting, filtering and other utilities. The main object in Pandas to contain data is the _Dataframe_

We can create a dataframe for example from:

* lists (una por fila) 
* diccionaries (key= column name, value= column (list))

```python
import pandas as pd

datos =[['Alphabet',1295],['Amazon',195],['Microsoft',952],['Apple',95]]
df = pd.DataFrame(datos, columns =['Company','Value'] )
df
```

From a dictionary:

```python
dat = {'Compañía':['Alphabet','Amazon','Microsoft'],'Value':[1250,876,567]}
```

```python
df2 = pd.DataFrame(dat)
df2
```

Also from a text file, for example a csv file:

```python
import pandas as pd
```

```python
# Read the CSV file
iris = pd.read_csv("iris.csv")

# View the first 5 rows
iris.head()
```

To acess a row you can use same notation as a dictionary:



```python
iris.iloc[1]
```

To access a column:

```python
iris['variety']
```

```python
iris['sepal.length'].iloc[0]
```

Its also possible to filter elements of a dataframe:

```python
iris['sepal.length']<3.0
```

```python
iris_greater7 = iris[iris['sepal.length']>7]
iris_greater7
```
